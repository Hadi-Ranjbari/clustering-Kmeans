# Feature selection

```python
y = new_df[['time.ppn']]
X = new_df.drop('time.ppn', axis=1)
print(X.columns[12] , X.columns[0])
print(X.shape, y.shape)
```

```python
print(X.isnull().sum())
print(y.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101, test_size=0.2)
```

```python
# we can use f score ANOVA to feature selection
fs = SelectKBest(k='all', score_func=f_classif)
fs.fit(X_train, y_train)
fs_X_train = fs.transform(X_train)
fs_X_test = fs.transform(X_test)
```

```python
# we can plot the each variable and the biiger is better
for i in range(len(fs.scores_)):
    plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
    plt.show()
    break
    
# location lusia & chol & ratio & hdl & glyhb
print(sorted(fs.scores_))
```
